```python
#3
%matplotlib inline
from scipy import fft
import matplotlib.pyplot as plt
import numpy as np


def func(x):
    return np.cos(x/15)+np.cos(x/2)

x = np.linspace(0, 256, 256)
plt.plot(x,func(x),color='k',linewidth=0.8)
plt.plot(x, np.cos(x/15), color = 'r', linewidth=0.8)
plt.plot(x, np.cos(x/2), color = 'b', linewidth=0.8)

plt.grid()
plt.show()



yy = func(x)
out = fft(yy)

def spectrum_power(x):
    re, im = x.real, x.imag
    return np.sqrt(re**2+im**2)

plt.plot(x,spectrum_power(out))
plt.xlim(0,128)
plt.show()

plt.plot(x,spectrum_power(out))
plt.xlim(0,128)
plt.yscale('log')
plt.show()
```
