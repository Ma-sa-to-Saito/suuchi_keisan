```python
%matplotlib inline
from scipy import fft
import matplotlib.pyplot as plt
import numpy as np


def func(x):
    return np.sin(x/13)+np.sin(x/2)


x = np.linspace(0, 306, 306)  #0から2πまでの範囲を100分割したnumpy配列
plt.plot(x, np.sin(x/13), color = 'r', linewidth=0.8)
plt.plot(x, np.sin(x/2), color = 'b', linewidth=0.8)

plt.grid()
plt.show()

plt.plot(x, func(x),color = 'k')
plt.xlim(0, 256)
plt.show()

yy = func(x)
out = fft(yy)

def spectrum_power(x):
    re, im = x.real, x.imag
    return np.sqrt(re**2+im**2)

plt.plot(x,spectrum_power(out))
plt.xlim(0,128)
plt.show()
```

```python
%matplotlib inline
from scipy import fft
import matplotlib.pyplot as plt
import numpy as np


def func(x):
    return 2*np.sin(x*2)+np.sin(x/2)


x = np.linspace(0, 306, 306)  #0から2πまでの範囲を100分割したnumpy配列
plt.plot(x, 2*np.sin(x*2), color = 'r', linewidth=0.8)
plt.plot(x, np.sin(x/2), color = 'b', linewidth=0.8)

plt.grid()
plt.show()

plt.plot(x, func(x),color = 'k')
plt.xlim(0, 256)
plt.show()

yy = func(x)
out = fft(yy)

def spectrum_power(x):
    re, im = x.real, x.imag
    return np.sqrt(re**2+im**2)

plt.plot(x,spectrum_power(out))
plt.xlim(0,128)
plt.show()

print('2sin(2x)+sin(x/2)の波長はsin(x/13)+sin(x/2)の波長と比べて波長が短くなっている。')
print('また周波数のピークはsin(x/13)は5の付近であるのに対し、2sin(2x)は100の付近である。')
```
