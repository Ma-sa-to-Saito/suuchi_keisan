```python
%matplotlib inline
from scipy import fft
import matplotlib.pyplot as plt
import numpy as np


def func(x):
    return np.sin(x/13)+np.sin(x/2)


x = np.linspace(0, 306, 306)  #02100numpy
plt.plot(x, np.sin(x/13), color = 'r', linewidth=0.8)
plt.plot(x, np.sin(x/2), color = 'b', linewidth=0.8)

plt.grid()
plt.show()

plt.plot(x, func(x),color = 'k')
plt.xlim(0, 256)
plt.show()

yy = func(x)
out = fft(yy)

def spectrum_power(x):
    re, im = x.real, x.imag
    return np.sqrt(re**2+im**2)

plt.plot(x,spectrum_power(out))
plt.xlim(0,128)
plt.show()
```
