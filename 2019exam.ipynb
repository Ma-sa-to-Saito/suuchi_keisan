```python
#1

import numpy as np
from pprint import pprint
import scipy.linalg as linalg

np.set_printoptions(precision=3, suppress=True)

#！！！！配列です！！！！変えてね！！！！
A = np.array([[5,-2,-2],[2,1,-2],[6,-2,-3]])
pprint(A)
#a = np.array([[3,2/3], [2/3,2]])
l,P = np.linalg.eig(A)
print('固有値')
pprint(l) #！！！固有値！！！
print('固有ベクトル')
pprint(P) #！！！固有ベクトル！！！

inv_P = linalg.inv(P)
print('対角化されているか')
pprint(np.dot(inv_P,np.dot(A,P)))
```

```python
#2 2016年過去問参照

#2-1
%matplotlib inline
from scipy import fft
import matplotlib.pyplot as plt
import numpy as np


def func(x):
    return np.cos(x/17)+np.cos(x/3)

print('黒が合成波')
x = np.linspace(0, 256, 256)
plt.plot(x,func(x),color='k',linewidth=0.8)
plt.plot(x, np.cos(x/17), color = 'r', linewidth=0.8)
plt.plot(x, np.cos(x/3), color = 'b', linewidth=0.8)

plt.grid()
plt.show()



yy = func(x)
out = fft(yy)

def spectrum_power(x):
    re, im = x.real, x.imag
    return np.sqrt(re**2+im**2)

plt.plot(x,spectrum_power(out))
plt.xlim(0,128)
plt.show()

plt.plot(x,spectrum_power(out))
plt.xlim(0,128)
plt.yscale('log')
plt.show()
```
